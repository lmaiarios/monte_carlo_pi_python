# Monte Carlo Algorithm for $\pi$ Estimation in Python

Esse notebook fará uma demonstração de como estimar $\pi$ usando um algoritmo de Monte Carlo.

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

Escolhemos o número de pontos:

```python
n_points = 1000
```

Os selecionamos aleatoriamente:

```python
x_data = np.random.rand(n_points)
y_data = np.random.rand(n_points)
```

Fazemos a contagem de quantos pontos há dentro da circunferência:

```python
n_circle = 0
for i in range(0,n_points-1):
    if (math.sqrt(x_data[i]**2+y_data[i]**2) <= 1):
        n_circle = n_circle+1
```

Portanto, nossa aproximação de $\pi$ é:

```python
aprox_pi = 4*(n_circle/n_points)
print(aprox_pi)
```

O erro percentual é:

```python
100*math.fabs(math.pi-aprox_pi)/math.pi
```

```python
circle_x = x_data[np.sqrt(x_data**2+y_data**2)<=1]
circle_y = y_data[np.sqrt(x_data**2+y_data**2)<=1]
```

```python
fig = plt.figure()
plot = fig.add_subplot(111)
plot.scatter(x_data, y_data, marker='.', color='blue')
plot.scatter(circle_x,circle_y, marker='.', color='red')

x = np.linspace(0, 1, 100)
y = np.sqrt(1-x**2)

plot.plot(x,y,color='black')

plot.set_aspect(1.0)

plt.rcParams['figure.figsize'] = [10, 10]
plt.show()
```
